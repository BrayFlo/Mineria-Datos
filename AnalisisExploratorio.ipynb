```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("streaming_platform.csv")

df.info()
```

```python
df.head()
```

```python
#hours_watched_per_week vs number_of_sessions

plt.figure(figsize=(6,4))
plt.scatter(data=df, x='hours_watched_per_week', y='number_of_sessions')
plt.title('Horas de vistas por semana vs Numero de sesiones')
plt.xlabel('Horas de vistas por semana')
plt.ylabel('Nunero de sesiones')
plt.show
```

```python
#age vs hours_watched_per_week

plt.figure(figsize=(6,4))
plt.scatter(data=df, x='age', y='hours_watched_per_week')
plt.title('Edad vs Horas Vistas por Semana')
plt.xlabel('Edad')
plt.ylabel('Horas Vistas por Semana')
plt.show
```

```python
#age vs account_creation_year

plt.figure(figsize=(6,4))
plt.scatter(data=df, x='age', y='account_creation_year')
plt.title('Edad vs Ano en que que se creo la cuenta')
plt.xlabel('Edad')
plt.ylabel('Ano en que se creo la cuenta')
plt.show
```

```python
#Pearson para las 3
pearson_hours_sessions = df['hours_watched_per_week'].corr(df['number_of_sessions'], method='pearson')
pearson_age_hours = df['age'].corr(df['hours_watched_per_week'], method='pearson')
pearson_age_account = df['age'].corr(df['account_creation_year'], method='pearson')

print ("Pearson: \n")
print ("Horas de vistas por semana vs Numero de sesiones: ", pearson_hours_sessions)
print ("Edad vs Horas Vistas por Semana: ", pearson_age_hours)
print ("Edad vs Ano en que que se creo la cuenta: ", pearson_age_account)
```
