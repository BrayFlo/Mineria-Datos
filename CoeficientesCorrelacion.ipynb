```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('datos_estudiantes_decision.csv')

df.info()
```

```python
df.head()
```

```python
#horas_estudio vs calificacion

plt.figure(figsize=(6,4))
plt.scatter(data=df, x='horas_estudio', y='calificacion')
plt.title('Horas de Estudio vs Calificacion')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificacion')
plt.show
```

```python
#horas_sueno vs nivel_estres
plt.figure(figsize=(6,4))
plt.scatter(data=df, x='horas_sueno', y='nivel_estres')
plt.title('Horas de Sueno vs Nivel de estres')
plt.xlabel('Horas de Sueno')
plt.ylabel('Nivel de estres')
plt.show
```

```python
#uso_redes vs calificacion
plt.figure(figsize=(6,4))
plt.scatter(data=df, x='uso_redes', y='calificacion')
plt.title('Uso de Redes vs Calificacion')
plt.xlabel('Uso de Redes')
plt.ylabel('Calificacion')
plt.show
```

```python
#Pearson para las 3
pearson_estudio_calf = df['horas_estudio'].corr(df['calificacion'], method='pearson')
pearson_sueno_estres = df['horas_sueno'].corr(df['nivel_estres'], method='pearson')
pearson_redes_calif = df['uso_redes'].corr(df['calificacion'], method='pearson')

print ("Pearson: \n")
print ("Horas de Estudio vs Calificacion: ", pearson_estudio_calf)
print ("Horas de Sueno vs Nivel de estres: ", pearson_sueno_estres)
print ("Uso de Redes vs Calificacion: ", pearson_redes_calif)
```

```python
#Spearman para las 3
spearman_estudio_calf = df['horas_estudio'].corr(df['calificacion'], method='spearman')
spearman_sueno_estres = df['horas_sueno'].corr(df['nivel_estres'], method='spearman')
spearman_redes_calif = df['uso_redes'].corr(df['calificacion'], method='spearman')

print ("Spearman: \n")
print ("Horas de Estudio vs Calificacion: ", spearman_estudio_calf)
print ("Horas de Sueno vs Nivel de estres: ", spearman_sueno_estres)
print ("Uso de Redes vs Calificacion: ", spearman_redes_calif)
```
